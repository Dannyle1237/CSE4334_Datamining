```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'sepal length (cm)': 's_length', 'sepal width (cm)': 's_width','petal length (cm)': 'p_length', 'petal width (cm)': 'p_width'}, inplace=True)
```

```python
features = df[['p_length', 'p_width', 's_length', 's_width']]
scaled_features = StandardScaler().fit_transform(features)
```

```python
pd.DataFrame(scaled_features)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_features)
principalComponents
```

```python
pcasolo = PCA(n_components = 1)
solo_PrincipalComponent = pcasolo.fit_transform(scaled_features)
```

```python
solo_PrincipalComponent
```

# Another way

```python
import pylab as pl
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
X_r
```

```python
# Percentage of variance explained for each components
print (pca.explained_variance_)
```

```python
pl.figure()
for c, i, target_name in zip("rgb", [0, 1, 2], target_names):
   pl.scatter(X_r[y==i,0], X_r[y==i,1], c=c, label=target_name)
pl.legend()
pl.title('PCA of IRIS dataset')

pl.show()
```

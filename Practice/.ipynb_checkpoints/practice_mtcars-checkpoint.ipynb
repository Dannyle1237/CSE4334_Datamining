```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Loading mtcars dataset and printing information about it
data = pd.read_csv('mtcars.csv')
print(data.head())
print(data.info())
print(data.describe())
```

```python
data.corr(method='pearson')
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(data.corr(method='pearson'),annot = True);
 
plt.show()
```

```python
#Simple regression

#x = data.iloc[:, 2].values   #uses only the cyl column use for model 2

#Adds a constant column, otherwise the regression does not calculate the intercept
X=sm.add_constant(x)
```

```python
#Multiple regression
#Change data to the right format
x2 = data.iloc[:, 2:].values

#Adds a constant column, otherwise the regression does not calculate the intercept
X2=sm.add_constant(x2)
```

```python
#reshaping y, otherwise matrix multiplication doesn't work
y = data.iloc[:, 1].values.reshape(-1,1)
```

```python
#check your X (exploratory variable array)
#X2
```

```python
model1=sm.OLS(y,X).fit()
model1.summary(xname=['intercept','cyl'])
```

```python
model1.mse_resid
```

#### Note: the above output already gives you the 95% confidence intervals. Try to change it to 99% confidence interval and see what happens!

```python
model1.conf_int(alpha=0.05, cols=None)
```

```python
model2=sm.OLS(y,X2).fit()
model2.summary(xname=['intercept','cyl', 'disp','hp', 'drat', 'wt', 'qsec','vs', 'am', 'gear', 'carb'])
```
